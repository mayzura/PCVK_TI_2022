```python
import cv2 as cv 
from google.colab.patches import cv2_imshow 
from skimage import io 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
print(' Gamma Correction pada citra ')
print('----------------------------------')
try:
 gamma = int(input('Masukkan nilai Gamma : '))
except ValueError:
 print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK_TI_2022/female.tiff')
gamma_image = np.zeros(original.shape, original.dtype)

#akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
    #memasukkan rumus gamma correction
     gamma_image[y,x,c] = np.clip(255 * pow(original[y,x,c] / 255, (1 / gamma)), 0, 255)
     
final_frame = cv.hconcat((original, gamma_image))
cv2_imshow(final_frame)
```

```python
print(' Simulasi Image Depth ')
print('----------------------------------')
try:
 bit_depth = float(input('Masukkan nilai Bit Depth : '))
except ValueError:
 print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK_TI_2022/rose_pink.png')
bit_depth_image = np.zeros(original.shape, original.dtype)

level = np.clip(255 / pow(2,bit_depth) - 1, 0, 255)

#akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
    #memasukkan rumus bit depth
     bit_depth_image[y,x,c] = np.clip(round((original[y,x,c] / level) * level), 0, 255)
     
final_frame = cv.hconcat((original, bit_depth_image))
cv2_imshow(final_frame)
```
